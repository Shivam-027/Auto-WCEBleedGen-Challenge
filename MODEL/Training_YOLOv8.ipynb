```python
import cv2
import numpy as np
from PIL import Image

from ultralytics import YOLO

# Load the YOLOv8 model
model = YOLO("runs/detect/train/weights/best.pt")

# Read the image
image = cv2.imread("/content/drive/MyDrive/WCEbleedGen/YOLO8/valid/images/img- (1305).png")

# Make a prediction
results = model.predict(image)

# Draw the predicted bounding boxes on the image
for result in results:
    if len(result) == 3:
        label, confidence, bbox = result
        cv2.rectangle(image, (bbox[0], bbox[1]), (bbox[2], bbox[3]), (0, 255, 0), 2)
        cv2.putText(image, label, (bbox[0], bbox[1] - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
    else:
        print(f"Warning: Unexpected result from YOLOv8 model: {result}")


# Display the image
cv2.imshow("Predicted Bounding Boxes", image)
cv2.waitKey(0)

# Print the label and bounded box
for result in results:
    label, confidence, bbox = result
    print(f"Label: {label}, Confidence: {confidence}, Bounding Box: {bbox}")
```
